Pandas란?

Pandas는 Python에서 데이터를 쉽게 다룰 수 있도록 도와주는 라이브러리야.

특히 표 형태의 데이터(엑셀, CSV, 데이터베이스 등)를 다룰 때 강력해.


왜 사용할까?

데이터 로딩: CSV, 엑셀, 데이터베이스에서 데이터를 쉽게 불러올 수 있어.

데이터 조작: 필터링, 정렬, 그룹화 등 원하는 방식으로 데이터를 변형할 수 있어.

데이터 분석: 평균, 합계, 분포 등 다양한 통계 연산이 가능해.

데이터 시각화: Matplotlib, Seaborn 같은 라이브러리와 함께 사용하면 그래프도
쉽게 만들 수 있어.

```python
import pandas as pd
import numpy as np  # numpy도 import 필요

# 날짜 생성
dates = pd.date_range("20130101", periods=6)

# 데이터프레임 생성
df = pd.DataFrame(np.random.randn(6, 4), index=dates, columns=list("ABCD"))

print(df)
```

```python
df2 = pd.DataFrame(
    {
        "A" : 1.0,
        "B" : pd.Timestamp("20130102"),
        "C" : pd.Series(1, index=list(range(4)), dtype="float32"),
        "D" : np.array([3] * 4, dtype= "int32"),
        "E" : pd.Categorical(["test", "train", "test", "train"]),
        "F" : "foo"

    }

)

df2
```

왜 이 코드의 결과값으로 나오는 행 개수는 4개인가?

왜냐면 C,D,E 를 코딩하는 부분에서 행이 4개로 정해졌기 때문이다. 나머지는 행의 개수만큼  수를 채워넣는 것이다. 그래서 결과로 행 4개가 나왔다.

Merge


```python
df = pd.DataFrame(np.random.randn(10, 4))
df
```

```python
pieces = [df[:3], df[3:7], df[7: ]]
pd.concat(pieces)
```

```python
left = pd.DataFrame({"key" : ["foo1", "foo2"], "lval" : [1, 2]})
right = pd.DataFrame({"key" : ["foo1", "foo2"], "rval" : [4, 5]})
print(left)
print(right)
```

```python
pd.merge(left, right, on="key")
```

Grouping


```python
df = pd.DataFrame(
    {
        "A" : ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B" : ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C" : np.random.randn(8),
        "D" : np.random.randn(8),

    }

    )

df
```

```python
df.groupby("A")[["C", "D"]].sum()
```

```python
df.groupby(["A", "B"]).sum()
```
